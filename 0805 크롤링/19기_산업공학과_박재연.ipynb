라이브러리 import

```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
actors_info_list = list()
for i in range(1,11):
    cine21_url = 'http://www.cine21.com/rank/person/content'
    post_data = dict()
    post_data['section'] = 'actor'
    post_data['period_start'] = '2021-07'
    post_data['gender'] = 'all'
    post_data['page'] = i
    res = requests.post(cine21_url, data=post_data)
    # post request (10번을 request한다)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    
    actors = soup.select('li.people_li')
    for actor in actors:
        actor_info_dict = dict()
        actor_link = 'http://www.cine21.com' + actor.select_one('div.name a')['href']
        actor_movies_link = actor_link.replace('info/?','info/?href=filmography&')
        
        actor_info_dict["이름"]=re.sub('\(\w*\)', '',actor.select_one('div.name').text)
        actor_info_dict["흥행지수"]=int(actor.select_one('strong').text.replace(',',''))
        
        #배우 출연영화
        actor_movies_list=list()
        response_actor_movies = requests.get(actor_movies_link)
        soup_actor_movies = BeautifulSoup(response_actor_movies.text, 'html.parser')
        actor_movies = soup_actor_movies.select('ul.works_history span.tit')
        for movie in actor_movies:
            actor_movies_list.append(re.sub('\(\w*\)', '', movie.text).strip())
        actor_info_dict["출연영화"]=actor_movies_list
        # 해당 배우의 모든 출연영화를 담기 위해
        # http://www.cine21.com/db/person/info/?href=filmography&person_id= 꼴의 페이지를 request했다
        
        actor_info_dict["랭킹"]=int(actor.select_one('span.grade').text)

        #배우 상세 페이지
        response_actor = requests.get(actor_link) # 배우 상세 페이지 이동
        soup_actor = BeautifulSoup(response_actor.text, 'html.parser')
        default_info = soup_actor.select_one('ul.default_info')
        actor_details = default_info.select('li')
    
        for actor_detail in actor_details:
            actor_detail_key = actor_detail.select_one('span.tit').text # 상세정보 key값
            actor_detail_value = re.sub('<span.*>.*</span>', '', str(actor_detail)) # span 태그 제거
            actor_detail_value = re.sub('<.*?>', '', actor_detail_value) # 상세정보 value값
            actor_info_dict[actor_detail_key] = actor_detail_value
        
        actors_info_list.append(actor_info_dict)
        
actors_info_list
```

```python
print(" 총 리스트 길이 :",len(actors_info_list))
```
